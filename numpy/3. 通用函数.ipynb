## 1. np.all
```
numpy.all(a, axis=None, out=None, keepdims=<no value>, *, where=<no value>)
```
+ a: array_like 类数组

+ axis: 轴无或整数或整数元组，可选参数

    执行逻辑与归约所沿的一个或多个轴。默认值（axis=None）是对输入数组的所有维度执行逻辑与运算。axis可以为负数，在这种情况下，它从最后一个轴开始向前计数。如果它是一个整数元组，则会在多个轴上执行归约，而不是像之前那样在单个轴或所有轴上执行。

+ keepdimsbool, optional

    用于放置结果的备用输出数组。它必须与预期输出具有相同的形状，并且其类型会被保留。

## 2. np.any

测试沿给定轴的任何数组元素是否为True

```python3
numpy.any(a, axis=None, out=None, keepdims=<no value>, *, where=<no value>)
```

+ a: 类数组

    可以转换为数组的输入数组或对象。

+ axis: None、整数或整数元组，可选参数

    执行逻辑“或”归约所沿的轴。默认值（axis=None）是对输入数组的所有维度执行逻辑“或”运算。axis可以为负数，在这种情况下，它从最后一个轴开始向前计数。如果它是一个整数元组，则会在多个轴上执行归约，而不是像之前那样在单个轴或所有轴上执行。

+ out

    用于放置结果的备用输出数组。它必须与预期输出具有相同的形状，并且其类型会被保留（例如，如果它是float类型，那么无论a的类型如何，它都将保持该类型，True返回1.0，False返回0.0）

+ keepdims

    如果将其设置为True，被缩减的轴将作为大小为1的维度保留在结果中。使用此选项，结果将与输入数组正确广播。

    如果传递了默认值，那么keepdims不会传递给ndarray子类的any方法，但任何非默认值都会被传递。如果子类的方法没有实现keepdims，则会引发任何异常

+ where

    检查任何True值时应包含的元素
